```python
from ultralytics import YOLO

model = YOLO('yolo11n.pt') 

# Start training
results = model.train(
    data=r"E:\German traffic signal classification\YOLO_Dataset\data.yaml", 
    epochs=80,      
    imgsz=640,       
    batch=2,        
    device=0,        
    workers=4,       
    augment=True,    
    name='gtsdb_rtx3050'
)
```

```python
import os
from ultralytics import YOLO

model_path = r"C:\Users\Asus\runs\detect\gtsdb_rtx30507\weights\best.pt"

if os.path.exists(model_path):
    print("Found the weights! Starting detection...")
    my_model = YOLO(model_path)
    
    video_path = r"E:\German traffic signal classification\traffic signal classification data\traffic-sign-to-test.mp4"
    
    results = my_model.predict(
        source=video_path, 
        save=True, 
        conf=0.3, 
        device=0
    )
else:
    print(f"Error: Could not find {model_path}. Please check if the training finished at least one epoch.")
```

```python
import cv2
from IPython.display import display, Image, clear_output
from ultralytics import YOLO

model_path = r"C:\Users\Asus\runs\detect\gtsdb_rtx30507\weights\best.pt"
my_model = YOLO(model_path)
video_path = r"E:\German traffic signal classification\traffic signal classification data\traffic-sign-to-test.mp4"

cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    success, frame = cap.read()
    if not success:
        break
        
    results = my_model(frame, device=0, verbose=False)
    annotated_frame = results[0].plot()
    
    _, buffer = cv2.imencode('.jpg', annotated_frame)
    clear_output(wait=True)
    display(Image(data=buffer.tobytes()))
    
cap.release()
print("Playback finished.")
```
